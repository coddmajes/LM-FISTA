```python
import numpy as np
import tensorflow as tf
import pathlib
import matplotlib.pyplot as plt
# from matplotlib.font_manager import _rebuild
from matplotlib.font_manager import FontProperties
from matplotlib.ticker import MultipleLocator

# _rebuild() #reload一下

from time import time
print(tf.__version__)
tf.keras.backend.set_floatx('float32')

makers = ["+--","+--","+--","+--","+--","+--","+--","+--","+--","+--","+--","+--",]
#makers = ["o--","<--","X--","s--","P--","o--","<--","X--","o--","<--","P--","x-","H-","P-"]
#makers = ["o--","<--","X--","s--","*--","P--","P--","x-","H-","P-"]
#makers = ["o--","<--","o--","<--","o--","<--","*--","P--","P--","x-","H-","P-"]
#co_p = [ 'red', 'red',"blue","blue","pink","purple","palegreen","camel"]
#co_p = ["purple",'green','blue', "blue","blue","pink","purple","palegreen","camel"]
#co_p = ['sandybrown', "white","white"]
#co_p = ['purple', 'sandybrown','sandybrown','steelblue','darkseagreen',]
#co_p = ['c', 'brown','green','purple',"navy","sandybrown","black"]

co_p = ['sandybrown', 'steelblue',"palevioletred","darkseagreen","gray","purple","blue","pink","purple","yellow","black"]
#co_p = ['purple', 'sandybrown','sandybrown','sandybrown','sandybrown',"green",'darkseagreen','blue','steelblue','steelblue',
        #"palevioletred",'palevioletred',"gray","gray","purple","blue","blue","pink","purple","palegreen","camel"]
#co_p = ['sandybrown', 'sandybrown','steelblue', 'steelblue',"darkseagreen","darkseagreen","gray","gray","purple","blue","blue","pink","purple","palegreen","camel"]
#co_p = ['sandybrown', 'sandybrown','sandybrown','steelblue','steelblue','steelblue', "darkseagreen",'darkseagreen','darkseagreen',"palevioletred","gray","gray","purple","blue","blue","pink","purple","palegreen","camel"]
#co_p = ['green', 'blue', "purple","pink","gray","gray","purple","blue","blue","pink","purple","palegreen","camel"]
title1, title2, title3, title4 = ["LISTA", "LISTA_cp", "TSS_LISTA_cp", "TSS_UISTA"]
test_titles = [title1, title2, title3, title4]
test_titles = ["Batch_Size_64","Batch_Size_256","Batch_Size_512","Batch_Size_1024", "Batch_Size_2048"]
layers = 16
colum=2
fsize=10
ymax=10
ymin=10
dire='upper right'
figsiz=(6,3)
ms=8
yspace=10
ylabel = 'NMSE(dB)'


def plot_contrast(in_x, in_tit):
    for i, mak, tit, x_,co_ in zip(np.arange(len(test_titles)), makers, test_titles, in_x,co_p):
        font = FontProperties(fname='/usr/share/fonts/truetype/arphic/uming.ttc')
        plt.plot(np.arange(layers), x_, mak,  color= co_,markersize=10,label=tit,linewidth=1.5) #mew=1,linewidth=1,
        #plt.plot(np.arange(layers), x_[0:10], mak,  color= co_,markerfacecolor= 'none',label=tit) #mew=1,linewidth=1,
        plt.xticks(np.arange(layers))
        plt.rcParams['figure.figsize'] = figsiz
        plt.tick_params(direction='in')
        plt.grid(ls='--')
        plt.ylim(ymin=ymin)
        plt.ylim(ymax=ymax)
        plt.xlabel('Iterations/Layers(k)',fontsize=12)
        #plt.xlabel(u'迭代次数/层数',fontproperties=font)
        plt.ylabel(ylabel,fontsize=12)
        #plt.ylabel('MSE')
        plt.title(in_tit)
        #plt.legend(ncol=colum, loc= 'lower left',fontsize=fsize)
        plt.legend(ncol=colum, loc= dire,fontsize=fsize, frameon=True, edgecolor='black',fancybox=False)
        #plt.legend(loc= 'upper right')
    plt.show()


def plot_d(in_x, label):
    for tit, mak, x_,co_ in zip(test_titles, makers, in_x,co_p):
        font = FontProperties(fname='/usr/share/fonts/truetype/arphic/uming.ttc')
        plt.plot(2+np.arange(layers), [x for x in x_], mak, color = co_, markerfacecolor= 'none',label=tit) #mew=1,linewidth=1,
        plt.xticks(2+np.arange(layers))
        
        plt.rcParams['figure.figsize'] = figsiz
        #plt.xlabel('Number of iterations or layers (k)')
        plt.xlabel(u'迭代次数/层数（k）',fontproperties=font)
        plt.ylim(ymin=ymin)
        plt.ylim(ymax=ymax)
        plt.ylabel(label,fontsize=16)
        plt.legend(ncol=colum, loc= dire,fontsize=fsize, frameon=True, edgecolor='black',fancybox=False)
        #plt.legend(ncol=1, loc= 'upper right',fontsize=9)
    plt.show()


def plot_(in_x, label):
    font = FontProperties(fname='/usr/share/fonts/truetype/arphic/uming.ttc')
    plt.plot(2+np.arange(layers), [x for x in in_x], "o--", color = "sandybrown") #mew=1,linewidth=1,
    plt.xticks(2+np.arange(layers))
    plt.rcParams['figure.figsize'] = (6, 3)
    #plt.rcParams['figure.figsize'] = (4, 2)
    plt.xlabel('Number of iterations or layers (k)')
    #plt.xlabel(u'迭代次数/层数(k)', fontproperties=font)
    plt.ylabel(label)
    #plt.legend(ncol=1, loc= 'upper right',fontsize=9)
    plt.show()
```
